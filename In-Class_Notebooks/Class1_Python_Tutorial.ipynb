## 3.2 Conditional statements: if, elif, else

Sometimes we want to condition an output on a given value (or similar scenario).

We can do this with statements using the if, elif (which stands for else if) or else commands.

NOTICE THE SYNTAX AND SPACING! The indent space is important!


```python
# let's assign some variables that hold some values

#this variable holds the age of our sun in years
sun_age = 4.6e9 #this is python for 4.6 x 10^9, or, 4,000,000,0000

#this variable holds the age of alpha centauri in years
alpha_cent_age = 4.8e9

#now let's print a sentence that depends on how the age of our sun compares to the age of alpha centauri
#we will use if, elif and else. Notice the use of the colon and the indentation!

### LET'S DESCRIBE THE SYNTAX TOGETHER!!
if sun_age > alpha_cent_age:
  print('Our sun is older than alphpa centauri.')
elif sun_age < alpha_cent_age:
  print('Our sun is younger than alpha centauri.')
else:
  print('unable to determine')
```

There are endless applications of conditional statements. They take the signs:

* $>$ (greater than)
* $<$ (less than)
* $>=$ (greater than or equal to)
* $<=$ (less than or equal to)
* == (is equal to)
* != (is not equal to)







## 3.3 Loops

Loops are very useful in writing scripts when we want to iterate through many items or values

#### for loops are the most common types of loops

```python
my_array
```

```python
galalxy_list
```

```python
# a "for loop" iterates through items in a list, tuple or array

# NOTICE THE SYNTAX!!
# Start with the key word "for"
# followed by a variable that you name, in this case we name our variable "i"
# then the key word "in"
# and then the array, list or other object you want to iterate through, in this case, "my_array" and a colon, :
# We then write the commands for waht we want to do. Here we just want to print the value of i
for i in my_array:
  print(i)

#another example
for j in galalxy_list:
  print(j)
```

```python
# python has a built-in range() function that is useful for iterating through integers

#notice the first and last integers printed
### WHAT IS HAPPENING IN THIS LOOP??
for i in range(5):
  print(i)
```

```python
for j in range(7,10):
  print(j)
```

```python
for k in range(5):
  print(k*2)
```

The built-in enumerate() function is particularly useful in for loops because not only does it iterate through elements in an array (or other object) ist also keeps count of the iteration.

```python
#WHAT'S HAPPENING IN THESE LINES OF CODE?

#this line does two things:
#it "enumerates"- counts the number of elements and identifies the elements in a list [2,4,6,8,10]
for p, cute in enumerate([2,4,6,8,10]):
  #this prints, the index of the element, the value of the element, and the element*10
  print(p, cute, cute*10)
```

This is useful for re-assigning values in an array

```python
# np.zeros generates an array of zeros
# we can re-assign the values of the zero array with values calculated through a loop like the following:

# let's designate a variable that will hold an array of 20 double values
# I will start with 20 zeros
doubles = np.zeros(20, dtype=float)

#now I will create a loop to calculate the doubles of 1-20
for i, el in enumerate(range(1,21)):
  doubles[i] = el*2

#check the new values of doubles
doubles
```

## 3.4 Functions

Writing functions are one of the MOST USEFUL tools in coding. Functions allow us re-use blocks of the same code as many times as needed.

```python
type(my_age)
```

```python
np.sqrt(100)
```

```python
# Functions are initialized with the key word def, followed by:
# the name of our function, the convention is to use lowercase letters and underscores,
# parentheses which contain the arguments to be passed to the function and a colon.
# Functions can also contain a """docstring""" which explains what the function does.

def force_func(mass, acceleration):
  '''
  This function calculates force (in Newtons) of a mass, with units in kg, accelerated with units m/s^2.

  parameters:
  -----------
  mass: (int, float or array of ints or floats)
  acceleration: (int, float or array of ints or floats)

  returns:
  --------
  force: (int, float or array of ints or floats)
  '''
  F = mass * acceleration

  #Here we use the return command.
  #Using return as opposed to print will allow us to assign a variable which stores the output of this function
  return F
```

```python
force_func(40.5, 2)
```

```python
#lets assign a variable that stores the force of a 50kg mass accelrated at 9.8m/s^2
f_test = force_func(50, 9.8)
```

# 4. $\underline{\mathrm{Basics\ of\ plotting\ with\ Matplotlib}}$

```python
# let's import the matplotlib package (library) and give it the shortcut plt

import matplotlib.pyplot as plt
```

```python
# let's create some x and y values to pass to our plotting function
x = np.linspace(0,50, num=50) #these will be our x-values, (independent variable)
y = x+1 #these will be our y-values (dependent variable)
```

Now let's make some standard plots

```python
# This creates a line plot
# you pass the values of the x-axis first, then the values of the y-axis as follows

### LET"S DESCRIBE THE SYNTAX AND OUTPUT TOGETHER
plt.plot(x,y)
```

```python
# we can control the appearance of this line
plt.plot(x,y, color='r', linestyle='dashed', linewidth=3)
```

matplotlib linestyles can be found here: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

matplotlib colors can be found here: https://matplotlib.org/stable/gallery/color/named_colors.html


```python
# if we want to compare may lines, we plot them one after the other in a cell like this
plt.plot(x,y,   color='r') #not specifying a linetsyle will produce a solid line
plt.plot(x,y+2, color='b', linestyle='dotted')
plt.plot(x,y+4, color='k', linestyle='dashed')
plt.plot(x,y+6, color='orange', linestyle='dashdot')
```

```python
# we can add axis labels to our figure also.

# notice the shortcut used for linstyles here
plt.plot(x,y,   color='r')
plt.plot(x,y+2, color='b', linestyle=':')
plt.plot(x,y+4, color='k', linestyle='--')
plt.plot(x,y+6, color='orange', linestyle='-.')

#we type our axis labels as strings
plt.xlabel('x-label [units]')
plt.ylabel('y-label [units]')
```

```python
# But how do we know what we're comparing??
# we can label each individual lina and call the built-in legend() function


#notice we now specify the label for each line
plt.plot(x,y,   color='r', label='x+1')
plt.plot(x,y+2, color='b', linestyle=':', label='x+3')
plt.plot(x,y+4, color='k', linestyle='--', label='x+5')
plt.plot(x,y+6, color='orange', linestyle='-.', label='x+7')

plt.xlabel('x-label [units]')
plt.ylabel('y-label [units]')

#and we must call the legend() function

plt.legend()
```

```python
# we can also add a title!

plt.plot(x,y,   color='r', label='x+1')
plt.plot(x,y+2, color='b', linestyle=':', label='x+3')
plt.plot(x,y+4, color='k', linestyle='--', label='x+5')
plt.plot(x,y+6, color='orange', linestyle='-.', label='x+7')

plt.xlabel('x-label [units]')
plt.ylabel('y-label [units]')

#let's give our plot a title
plt.title('Awesome title')

plt.legend()
```

```python
# We can do the same, but with a SCATTER plot.
# this is useful if we want to see individual datapoints

#The differences is now we use the plt.scatter function
#and instead of specifying the linestyle, we specify the marker

plt.scatter(x,y,   color='r', label='x+1')
plt.scatter(x,y+2, color='b', marker='.', label='x+3')
plt.scatter(x,y+4, color='k', marker='*', label='x+5')
plt.scatter(x,y+6, color='orange', marker='^', label='x+7')

plt.xlabel('x-label [units]')
plt.ylabel('y-label [units]')

#let's give our plot a title
plt.title('Awesome title')

plt.legend()
```
