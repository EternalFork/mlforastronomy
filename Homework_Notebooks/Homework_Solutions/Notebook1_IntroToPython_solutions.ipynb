# Problem 2: fun with escape velocity

Escape velocity is the velocity needed by an object to escape the gravitational force of a celestial body. The formula is:
$$ v_{\rm esc} = \sqrt{\frac{2GM}{r}}$$

where G is Newton's constant of gravitation (always represented by a capital G) experimentally determined with a value $6.67\times10^{-11}$ and units $\left[\frac{\rm m^3}{\rm kg\ s^2}\right]$

$M$ is the mass of the celestial body (such as a planet), and

$r$ is the radius of the same cestial body.

#### 2a. Write a function that takes mass and radius as arguments and returns escape velocity in units of meters per second, $[{\rm m/s}]$.
Be sure to include a docstring to explain your function.
You will want to be midful ofth eunits you wish to pass in order to output the correct velocity units.

```python
def v_esc(m,r):
    '''
    This function calculates the escape velocity of a celestial body.
    parameters:
    ----------
    m: (float, mass of body in kg)
    r: (float, radius of body in meters)

    returns:
    --------
    escape velocity: (float, excape velocity in m/s)
    '''
    G = 6.67e-11 #in units of m^3/(kg s^2)
    v = np.sqrt((2*G*m)/r)
    return v
```

```python
#this line creates a variable x1 that stores an array linearly spaced between 0.0-5.0
x1 = np.linspace(0.0, 5.0)
#this line creates a variable x2 that stores an array linearly spaced between 0.0-2.0
x2 = np.linspace(0.0, 2.0)

#this line creates a variable y1 that stores the output of the equation cos(2pix1)*exp(-1)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
#this line creates a variable y2 that holds the output of equation cos(2pix2)
y2 = np.cos(2 * np.pi * x2)

#this line initiates fig and ax variables which hold a matplotlib figure with two 2 rows and 1 column of subplot panels
fig, ax = plt.subplots(2, 1)

# this line gives the fugure a title
fig.suptitle('A tale of 2 subplots')

#this plots x1 vs y1 in the first panel with circles connected by a line
ax[0].plot(x1, y1, 'o-')
#this sets the y-axis label of the first panel
ax[0].set_ylabel('Damped oscillation')

#this plots x2 vs y2 in the second panel with small dos connected by a line
ax[1].plot(x2, y2, '.-')
#this sets the x-axis label
ax[1].set_xlabel('time (s)')
#this sets the y-axis label of the second panel
ax[1].set_ylabel('Undamped')
```
